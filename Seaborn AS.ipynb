```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

np.random.seed(42)
x = np.random.rand(100) * 100  # Variable X
y = 2.7* x + np.random.normal(0, 10, 100)  # Variable Y with noise

df_scatter = pd.DataFrame({'X': x, 'Y': y})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_scatter, x='X', y='Y', color='blue', alpha=0.6 )
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.



random_numbers = np.random.randn(1000)  # Normal distribution

df_distribution = pd.DataFrame({'Random Numbers': random_numbers})

plt.figure(figsize=(10, 6))
sns.histplot(df_distribution['Random Numbers'], bins=30, kde=True, color='orange', alpha=0.7)
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values



categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 20]
df_categories = pd.DataFrame({'Category': categories, 'Values': values})


plt.figure(figsize=(10, 6))
sns.barplot(data=df_categories, x='Category', y='Values', palette='viridis')
plt.title('Comparison of Categories Based on Values')
plt.ylabel('Values')
plt.grid(True)
plt.show()

```

```python

#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

np.random.seed(0)
df_category_values = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D'], 200),
    'Value': np.random.randn(200) * 10 + 50  # Normal distribution around 50
})

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_category_values, x='Category', y='Value', palette='Set2')
plt.title('Distribution of Values Across Different Categories')
plt.ylabel('Value')
plt.grid(True)
plt.show()

 

```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

data = np.random.rand(100, 5)
correlated_data = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])
correlated_data['B'] = correlated_data['A'] * 0.5 + np.random.normal(0, 0.1, 100)  # Correlation
correlated_data['C'] = correlated_data['A'] * 0.2 + np.random.normal(0, 0.1, 100)

# Calculate the correlation matrix
correlation_matrix = correlated_data.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()

```
